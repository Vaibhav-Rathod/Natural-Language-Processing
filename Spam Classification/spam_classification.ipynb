## Data cleaning and preprocessing

```python
#Data cleaning and preprocessing
import re
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
#creating object for stemming of words 
ps=PorterStemmer()
lem=WordNetLemmatizer()
```

```python
corpus=[]
for i in range(0,len(msg)):
    review=re.sub("[^a-zA-Z0-9]"," ",msg["Message"][i])
    review=review.lower()
    review=review.split()
    #creating stem by WordNetLemmatizer
    review=[lem.lemmatize(word) for word in review if not word in stopwords.words("english")]
    review =" ".join(review)

    corpus.append(review)
```

```python
#Messages after the lemmatization
corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
```

```python
X=tfidf.fit_transform(corpus).toarray()
```

```python
# Msg after preforming TFIDF Vectorizaition 
for i in range(len(X)):`
    print(X[i])
```

```python
y=pd.get_dummies(msg["Type"])
y=y.iloc[:,1].values
```

## Training Model

```python
#Spliting into train and test
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=0.25,shuffle=True)
```

```python
#Training model 
from sklearn.naive_bayes import MultinomialNB
spam_detect=MultinomialNB().fit(xtrain,ytrain)
```

## Testing 

```python
pred=spam_detect.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
cm=confusion_matrix(ytest,pred,labels=[0,1])
print(cm)
```

```python
cf=classification_report(ytest,pred,labels=[0,1])
print(cf)
```

```python
acc=accuracy_score(ytest,pred)
print(acc)
```

```python
class Solution(object):
    def uniquePaths(self, m, n):
        dyn=[[0]*n]*m
        for i in range(m):
            for j in range(n):
                if i==0 or j==0:
                    dyn[i][j]=1
                else :
                    dyn[i][j]=dyn[i-1][j]+dyn[i][j-1]
        return dyn[m-1][n-1]

m=int(input())
n=int(input())
sol=Solution()
print(sol.uniquePaths(m,n))
```
